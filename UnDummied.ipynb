```python
import pandas as pd
```

```python
predicted_data = pd.read_csv("Predicted_Data.csv")
predicted_data.head()
```

```python
zip_codes = ['46214', '46219', '46220', '46227']
```

```python
predicted1 = pd.DataFrame()
```

```python
for zip in zip_codes:
    df = predicted_data.loc[predicted_data[zip]==1]
    df = df.drop(zip,axis=1)
    df["Zip_Code"] = zip
    predicted1 = pd.concat([predicted1,df])
```

```python
predicted1["Zip_Code"]
```

```python
print(len(predicted1["Zip_Code"]))
```

```python
predicted1 = predicted1.drop(["46214","46219","46220","46227"], axis=1)
```

```python
predicted1.tail()
```

```python
print(len(predicted1["Zip_Code"]))
```

```python
bedroom = {"1 Bedroom": 1, "2 Bedroom": 2, "3 Bedroom": 3, "4 Bedroom": 4, "5 Bedroom": 5}
```

```python
predicted2 = pd.DataFrame()
```

```python
for key,value in bedroom.items():
    df = predicted1.loc[predicted1[key]==1]
    df["No. of Bedrooms"] = value
    predicted2 = pd.concat([predicted2,df])
    
```

```python
predicted2.tail()
```

```python
print(len(predicted2["No. of Bedrooms"]))
```

```python
predicted2 = predicted2.drop(["1 Bedroom", "2 Bedroom", "3 Bedroom", "4 Bedroom", "5 Bedroom"], axis=1)
```

```python
predicted2.head()
```

```python
print(len(predicted2["No. of Bedrooms"]))
```

```python
month = {"January":1, "February":2, "March":3, "April":4, "May":5, "June":6, "July":7, "August":8, "September":9,
        "October":10, "November":11, "December":12}
```

```python
predicted3 = pd.DataFrame()
```

```python
for key,value in month.items():
    df = predicted2.loc[predicted2[key]==1]
    df["Month"] = value
    predicted3 = pd.concat([predicted3,df])
```

```python
predicted3.tail()
```

```python
print(len(predicted3["Month"]))
```

```python
predicted3 = predicted3.drop(["January", "February", "March", "April", "May", "June", "July", "August", "September",
        "October", "November", "December"], axis=1)
```
